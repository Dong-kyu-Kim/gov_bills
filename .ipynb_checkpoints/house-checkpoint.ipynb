```python
import os
import pandas as pd
```

```python
house_rep_csv = os.path.join("csv_files", "House_rep.csv")
house_vote_tallies_csv = os.path.join("csv_files", "house_vote_tallies_modified.csv")

house_rep_df = pd.read_csv(house_rep_csv, encoding = 'utf8')
house_vote_tallies_df = pd.read_csv(house_vote_tallies_csv, encoding = 'utf8')
```

```python
house_rep_df.tail()
```

```python
house_vote_tallies_df.tail()
```

```python
filtered_house_rep_df = house_rep_df[house_rep_df["Phone"].isin(house_vote_tallies_df["Phone Number"])]
filtered_house_rep_df = filtered_house_rep_df[filtered_house_rep_df["Name"].isin(house_vote_tallies_df["Full Name"])]
filtered_house_rep_df.tail()
```

```python
output_csv_path = os.path.join("csv_files", "filtered_house_rep_df.csv")
filtered_house_rep_df.to_csv(output_csv_path, encoding = 'utf8', index = False)
```
