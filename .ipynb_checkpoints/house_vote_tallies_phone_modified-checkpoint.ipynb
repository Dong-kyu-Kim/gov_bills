```python
import os
import pandas as pd
```

```python
csv_path = os.path.join("csv_files", "house_vote_tallies_2.csv")
df = pd.read_csv(csv_path, encoding = 'utf8')
df.tail()
```

```python
df['Phone Number'] = df['Phone Number'].str.replace(r"[()]", "", regex=True).str.replace(" ", "-")
df['Phone Number'] = df['Phone Number'].str.lstrip("-")
df.tail()
```

```python
output_csv_path = os.path.join("csv_files", "house_vote_tallies_modified.csv")
df.to_csv(output_csv_path, encoding = 'utf8', index = False)
```
